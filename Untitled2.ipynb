```python
## RCCN 2

from base64 import decode
import http
import json
import urllib
from webbrowser import get

import requests


#vendas ingressos
url1 =  'https://us-central1-raccoon-bi.cloudfunctions.net/psel_de_ingressos'

handle = urllib.request.urlopen(url1)
vendas = handle.read().decode()
print('tamanho vendas' , len(vendas))
print('tipo vendas' , type(vendas))
print('vendas',vendas)
#print(vendas(1))
vend = json.loads(vendas)
print('vend' , vend)
print('tipo vend' , type(vend))

for i in range(0,len(vend)):
    print(vend[i]['tipo'])

#print(vend[1])




# Gastos ingressos 
url2 = 'https://us-central1-raccoon-bi.cloudfunctions.net/psel_de_compras'
gastos=[]
handle = urllib.request.urlopen(url2)
for data in handle:
    c = data.decode()
    lst = c.split(',')
    lst[2] = lst[2].replace('\r\n',"")
    gastos.append(lst)

print('tipo gastos: ', type(gastos))
print('gastos: ',len(gastos))
print('gastos: ', gastos)

for i in range(0,len(gastos)):
    print(gastos[i])
```
